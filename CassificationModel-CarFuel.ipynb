### Machine Learning Model: Classification of Car Fuel

Criar um modelo de classificacao usando regressao logistica para prever qual o tipo de combustivel do carro 

```python
import pandas as pd
import sklearn.metrics as metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df
```

```python
df.columns
```

```python
y = df['gas'].values
y
```

1: Gasolina
0: Diesel

```python
x = df[['symboling', 'normalized-losses', 'wheel-base', 'length',
       'width', 'height']].values
x.dtype
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
model = LogisticRegression(C = 0.7, random_state = 42, max_iter = 1000)
```

```python
model.fit(x_train,y_train)
```

```python
y_preds = model.predict(x_test)
```

```python
print('Acc:',metrics.accuracy_score(y_test,y_preds))
print('Confusion Matrix:\n',metrics.confusion_matrix(y_test,y_preds, labels = [0,1]))
print('Classification:\n',metrics.classification_report(y_test,y_preds))
```
